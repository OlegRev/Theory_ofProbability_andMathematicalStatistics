```python
import numpy as np
import pandas as pd
import scipy.stats
```

### 1) –î–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è –≤–µ–ª–∏—á–∏–Ω—ã –∑–∞—Ä–∞–±–æ—Ç–Ω–æ–π –ø–ª–∞—Ç—ã –∑–∞–µ–º—â–∏–∫–æ–≤ –±–∞–Ω–∫–∞ (zp) –∏ –∑–Ω–∞—á–µ–Ω–∏—è –∏—Ö –ø–æ–≤–µ–¥–µ–Ω—á–µ—Å–∫–æ–≥–æ –∫—Ä–µ–¥–∏—Ç–Ω–æ–≥–æ —Å–∫–æ—Ä–∏–Ω–≥–∞ (ks):
    
    zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
    ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].

–ù–∞–π–¥–∏—Ç–µ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—é —ç—Ç–∏—Ö –¥–≤—É—Ö –≤–µ–ª–∏—á–∏–Ω —Å –ø–æ–º–æ—â—å—é —ç–ª–µ–º–µ–Ω—Ç–∞—Ä–Ω—ã—Ö –¥–µ–π—Å—Ç–≤–∏–π,
–∞ –∑–∞—Ç–µ–º —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ cov –∏–∑ numpy
–ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ä–∞–≤–Ω—ã.

–ù–∞–π–¥–∏—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ —Å –ø–æ–º–æ—â—å—é –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ 
–∏ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –¥–≤—É—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤,

–∞ –∑–∞—Ç–µ–º —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ñ—É–Ω–∫—Ü–∏–π –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫ numpy –∏ pandas.

–§–æ—Ä–º—É–ª–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏:
    
$$cov_{XY} = M(XY) - M(X)M(Y)$$
–§–æ—Ä–º—É–ª–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞:    
$$r_{XY} = \frac{cov_{XY}}{\sigma_{X}\sigma_{Y}}$$


```python

zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
df = pd.DataFrame({'zp': zp, 'ks': ks}, columns=['zp', 'ks'])


zp_ks = zp*ks
cov_XY = zp_ks.mean() - zp.mean()*ks.mean()
D_zp = zp.var()
ùúé_zp = np.sqrt(D_zp)
D_ks = ks.var()
ùúé_ks = np.sqrt(D_ks)
r_XY = cov_XY/(ùúé_zp*ùúé_ks)

np_cov_XY = np.cov(zp,ks,ddof=0)[0, 1]
std_zp = zp.std()
std_ks = ks.std()
np_r_XY = np_cov_XY/(std_zp*std_ks)

pd_cov_XY = df.cov().values[0, 1]
pd_r_XY = df.corr()
print(f'ùëêùëúùë£ùëãùëå = ùëÄ(ùëãùëå)‚àíùëÄ(ùëã)ùëÄ(ùëå) : {cov_XY},\n'
      f'np.cov(zp,ks,ddof=0)[0, 1] : {np_cov_XY}\n\n'
      f'round(ùëêùëúùë£ùëãùëå, 2)==round(np_cov_XY, 2) :'
      f' {round(cov_XY, 2)==round(np_cov_XY, 2)}\n'
      f'–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ —Å –ø–æ–º–æ—â—å—é –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ \n'
      f'–∏ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –¥–≤—É—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ : {r_XY}\n'
      f'—Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ñ—É–Ω–∫—Ü–∏–π –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫ numpy :'
      f' {np.cov(zp,ks,ddof=0)[0, 1]/ (std_zp*std_ks)}\n'
      f'–∏ pandas :\n {pd_r_XY}')
```

### 2) –ò–∑–º–µ—Ä–µ–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è IQ –≤—ã–±–æ—Ä–∫–∏ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –æ–±—É—á–∞—é—â–∏—Ö—Å—è –≤ –º–µ—Å—Ç–Ω—ã—Ö —Ç–µ—Ö–Ω–∏—á–µ—Å–∫–∏—Ö –≤—É–∑–∞—Ö:
    
    131, 125, 115, 122, 131, 115, 107, 99, 125, 111.

–ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –≤ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ IQ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω –Ω–æ—Ä–º–∞–ª—å–Ω–æ.

–ù–∞–π–¥–∏—Ç–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è —Å –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç—å—é 0.95.

$$\alpha=0.05\quad \frac{\alpha}{2} = 0.025$$


```python
IQ = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
n = len(IQ)
M_IQ = IQ.mean()
ùõº = 0.05
ùõº_divided_by_2 = ùõº/2
degree_of_freedom = n - 1
t_Student = scipy.stats.t.ppf(1-ùõº_divided_by_2, degree_of_freedom)


M_minus = M_IQ - (t_Student * IQ.std(ddof=1))/np.sqrt(n)
M_plus = M_IQ + (t_Student * IQ.std(ddof=1))/np.sqrt(n)

print(f'–ó–Ω–∞—á–µ–Ω–∏—è IQ –≤—ã–±–æ—Ä–∫–∏ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤: {IQ}\n'
      f'–í–µ–ª–µ—á–∏–Ω–∞ –≤—ã–±–æ—Ä–∫–∏: {n}\n'
      f'M(IQ) = {M_IQ}\n'
      f'–ù–∞–¥–µ–∂–Ω–æ—Å—Ç—å—é = 0.95, ùõº = {ùõº}, ùõº/2 = {ùõº_divided_by_2}\n'
      f'–°—Ç–µ–ø–µ–Ω—å —Å–≤–æ–±–æ–¥—ã : {degree_of_freedom}\n'
      f't –∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞ : {t_Student}\n'
      f'–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è\n'
      f' —Å –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç—å—é 0.95 : [{M_minus}, {M_plus}]')
```


### 3) –ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ —Ä–æ—Å—Ç —Ñ—É—Ç–±–æ–ª–∏—Å—Ç–æ–≤ –≤ —Å–±–æ—Ä–Ω–æ–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω –Ω–æ—Ä–º–∞–ª—å–Ω–æ
—Å –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, —Ä–∞–≤–Ω–æ–π 25 –∫–≤.—Å–º. –û–±—ä–µ–º –≤—ã–±–æ—Ä–∫–∏ —Ä–∞–≤–µ–Ω 27,
—Å—Ä–µ–¥–Ω–µ–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 174.2. –ù–∞–π–¥–∏—Ç–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ
–æ–∂–∏–¥–∞–Ω–∏—è —Å –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç—å—é 0.95.




```python
D = 25
ùúé = np.sqrt(D)
n = 27
M = 174.2
ùõº = 0.05
ùõº_divided_by_2 = ùõº/2

z_t = 1.96
M_minus = M - (z_t * ùúé)/np.sqrt(n)
M_plus = M + (z_t * ùúé)/np.sqrt(n)

print(f'–í–µ–ª–µ—á–∏–Ω–∞ –≤—ã–±–æ—Ä–∫–∏: {n}\n'
      f'–°—Ä–µ–¥–Ω–µ–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç : {M}\n'
      f'–¥–∏—Å–ø–µ—Ä—Å–∏–µ–π –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, —Ä–∞–≤–Ω–æ–π {D} –∫–≤.—Å–º \n'
      f'ùúé = {ùúé}\n'
      f'–ù–∞–¥–µ–∂–Ω–æ—Å—Ç—å—é = 0.95, ùõº = {ùõº}, ùõº/2 = {ùõº_divided_by_2}\n'
      f'Z –∫—Ä–∏—Ç–µ—Ä–∏–π : {z_t}\n'
      f'–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è\n'
      f' —Å –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç—å—é 0.95 : [{M_minus}, {M_plus}]')
```
